```python
from tkinter import * 
from tkinter import Tk 
```

```python
root = Tk() 
root.title("Simple calculation")

number = Entry(root  , width=50, borderwidth=5)
number.grid(row=0 ,column = 0 , columnspan =3  , padx = 5  , pady = 5) 

#defining the function ss 
def clicknumber(numberr) :
    
    current = number.get()
    number.delete(0 , END )
    number.insert(0 , str(current) + str(numberr))
    
    

# defining the  clear function 

def clickclear():
    number.delete(0, END)
    
    
# definning  the add function
    
def clickadd(): 
    first_number  = number.get() 
    
    global f_num
    global math  
    math  = "addition"
    f_num =  float(first_number) 
    number.delete(0, END)
    
    
# defining the equal function 
def clickequal():
    last_number  = number.get() 
    number.delete(0 , END)
    global l_number 
    l_number  = float(last_number)
    
    if math == "addition":
        number.insert(0 , f_num + l_number)
    else: 
        if math == "soustraction":
            number.insert(0 , f_num - l_number)
        else: 
            if math  == "multiplication":
                number.insert(0 , f_num*l_number)
            else :
                number.insert(0 , f_num / l_number)
                
                
    
    
# defining the soustraction  function 

def clicksoustraction():
    first_number  = number.get() 
    
    global f_num
    global math  
    math  = "soustraction"
    f_num =  float(first_number) 
    number.delete(0, END)
# defining the multiplication function 
def clickmultiplicaition():
    first_number  = number.get() 
    
    global f_num
    global math  
    math  = "multiplication"
    f_num =  float(first_number) 
    number.delete(0, END)

#define the divition function 

def clickdivition():
    first_number  = number.get() 
    
    global f_num
    global math  
    math  = "divition"
    f_num =  float(first_number) 
    number.delete(0, END) 

    
    
    
    
    
    

#creating the button of the number using in our calculation 

button_1  =  Button(root ,text = "1" , padx = 40 ,  pady= 20  , command = lambda:clicknumber(1) )
button_2  =  Button(root ,text = "2" , padx = 40 ,  pady= 20  , command = lambda:clicknumber(2) )
button_3  =  Button(root ,text = "3" , padx = 40 ,  pady= 20  , command = lambda:clicknumber(3) )

button_4  =  Button(root ,text = "4", padx = 40 ,  pady= 20  , command = lambda:clicknumber(4))
button_5  =  Button(root , text = "5" ,padx = 40 ,  pady= 20  , command = lambda:clicknumber(5))
button_6  =  Button(root ,text = "6" , padx = 40 ,  pady= 20  , command = lambda:clicknumber(6))


button_7  =  Button(root ,text = "7" , padx = 40 ,  pady= 20  , command = lambda:clicknumber(7))
button_8  =  Button(root ,text = "8" , padx = 40 ,  pady= 20  , command = lambda:clicknumber(8))
button_9  =  Button(root ,text = "9" , padx = 40 ,  pady= 20  , command = lambda:clicknumber(9) )

button_0  =  Button(root ,text = "0" , padx = 40 ,  pady= 20  , command = lambda:clicknumber(0))
#adding the button + , = and clear 

button_add = Button(root  , text = "+" , padx = 40  , pady= 20  , command =clickadd )
button_equal  = Button(root  ,text = "=" ,  padx = 91  , pady= 20  , command =clickequal )
button_clear  = Button(root  , text = "C" ,padx = 91  , pady= 20  , command =lambda: clickclear() )

# add the others operations -  ,* , /  buttons 

button_soustraction  = Button(root  ,  text ="-" , padx = 40  , pady= 20, command = clicksoustraction )
button_multiply  = Button(root  ,  text ="*" , padx = 40  , pady= 20, command = clickmultiplicaition )
button_divide   = Button(root  ,  text ="/", padx = 40  , pady= 20 , command = clickdivition )


#shoving the buttons to the screens : 

button_1.grid(row = 3, column = 0)
button_2.grid(row = 3, column = 1)
button_3.grid(row = 3, column = 2)

button_4.grid(row =2 , column = 0)
button_5.grid(row =2 , column = 1)
button_6.grid(row =2 , column = 2)

button_7.grid(row = 1, column = 0)
button_8.grid(row = 1, column = 1)
button_9.grid(row = 1, column = 2)

button_0.grid(row =4 , column = 0)
button_add.grid(row =5 , column = 0 )
button_clear.grid(row =5 , column = 1 , columnspan = 2)
button_equal.grid(row =4 , column = 1, columnspan = 2)
#shoving the operation button to  the screen : 
button_soustraction.grid(row = 6,  column = 0)
button_multiply.grid(row = 6,  column = 1)
button_divide.grid(row = 6,  column = 2)



root.mainloop()
```
