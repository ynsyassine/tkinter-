```python
from tkinter import * 
from tkinter import Tk 
from PIL import Image, ImageTk
```

```python
root = Tk()
#add the title 

root.title("exit button and image")


#create the image   
myimage1  = ImageTk.PhotoImage(Image.open("/home/yassine/images/yassine1.jpg")) 
myimage2  = ImageTk.PhotoImage(Image.open("/home/yassine/images/yassine2.jpg"))
myimage3  = ImageTk.PhotoImage(Image.open("/home/yassine/images/yassine3.jpg"))
myimage4  = ImageTk.PhotoImage(Image.open("/home/yassine/images/yassine4.jpg"))

#create a list of images 
imglist = [myimage1 , myimage2  , myimage3 , myimage4 ] 
#create a label for myimage 
 
imglabel = Label(image=myimage2)

#shoving the image into screen 
imglabel.grid(row = 0 , column = 0 , columnspan = 3)

# creating a function in order destroy the root 
def exit_program():
   
    root.destroy()
    
#creating the function for the next image
def  next_one(image_number):
    global imglabel 
    global forward_button 
    global previous_button
    imglabel.grid_forget()
    imglabel = Label(image=imglist[image_number -1])
    forward_button = Button(root , text = ">>" , command = lambda:next_one(image_number+1))
    forward_button.grid(row = 1 , column = 2 )
    if image_number == 4 :
            forward_button.configure(state=DISABLED)

        
    previous_button = Button(root , text = "<<" , command = lambda:previous_one(image_number-1))
    previous_button.grid(row = 1  ,  column = 0)
    imglabel.grid(row = 0 , column = 0 , columnspan = 3)

    
    


        
    
#creating the function for the previous image
def previous_one(image_number):
    global imglabel 
    global forward_button 
    global previous_button
    imglabel.grid_forget()
    imglabel = Label(image=imglist[image_number - 1])
    imglabel.grid(row = 0 , column = 0 , columnspan = 3)
    forward_button = Button(root , text = ">>" , command = lambda:next_one(image_number+1))
    forward_button.grid(row = 1 , column = 2 )
    if image_number == 1 :
            previous_button.configure(state=DISABLED)

        
    previous_button = Button(root , text = "<<" , command = lambda:previous_one(image_number-1))
    previous_button.grid(row = 1  ,  column = 0)
    imglabel.grid(row = 0 , column = 0 , columnspan = 3)

    
    
    
        
    
     


    
#use the function to destroy the root 
forward_button = Button(root , text = ">>" , command = lambda:next_one(1))
exit_button = Button(root, text="Exit", command=exit_program)
previous_button = Button(root , text = "<<" , command = lambda:previous_one(1))

#put the button into the screen  : 

previous_button.grid(row = 1  ,  column = 0)
exit_button.grid(row = 1  , column = 1 )
forward_button.grid(row = 1 , column = 2 )


    
                  


root.mainloop()
```
